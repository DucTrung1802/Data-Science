```python
# We import everything that we'll use

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
questions = pd.read_csv("2019_questions.csv", parse_dates=["CreationDate"])
```

```python
questions.info()
```

```python
questions["Tags"].apply(lambda value: type(value)).unique()
```

```python
questions.fillna(value={"FavoriteCount": 0}, inplace=True)
questions["FavoriteCount"] = questions["FavoriteCount"].astype(int)
questions.dtypes
```

```python
questions["Tags"] = questions["Tags"].str.replace("^<|>$", "").str.split("><")
questions.sample(20)
```

```python
tag_count = {}

def count_tag(tags):
    for tag in tags:
        if tag in tag_count:
            tag_count[tag] += 1
        else:
            tag_count[tag] = 1

questions["Tags"].apply(count_tag)
```

```python
tag_count = pd.DataFrame.from_dict(tag_count, orient="index")
tag_count.rename(columns={0: "Count"}, inplace=True)
tag_count.sort_values("Count", ascending=False, inplace=True)
tag_count
```

```python
most_used = tag_count.sort_values(by="Count").tail(20)
most_used
```

```python
most_used.plot.barh()
```

```python
tag_view_count = dict()

for index, row in questions.iterrows():
    for tag in row['Tags']:
        if tag in tag_view_count:
            tag_view_count[tag] += row['ViewCount']
        else:
            tag_view_count[tag] = row['ViewCount']
            
most_viewed = pd.DataFrame.from_dict(tag_view_count, orient="index")
most_viewed.rename(columns={0: "view_count"}, inplace=True)
most_viewed.sort_values(by="view_count", inplace=True)
most_viewed_top_20 = most_viewed.tail(20)
most_viewed_top_20.plot.barh()
```

```python
in_used = pd.merge(most_used, most_viewed, how="left", left_index=True, right_index=True)
in_used.sort_values("Count", inplace=True, ascending=False)
in_used
```

```python
in_view = pd.merge(most_used, most_viewed, how="right", left_index=True, right_index=True)
in_view.sort_values("view_count", inplace=True, ascending=False)
in_view
```

```python
all_tags = list(tag_count.index)
all_tags
```

```python
associations = pd.DataFrame(index=all_tags, columns=all_tags)
associations.iloc[0:6,0:6]
```

```python
associations.fillna(0, inplace=True)

for tags in questions["Tags"]:
    associations.loc[tags, tags] += 1
```

```python
associations
```

```python
relations_most_used = associations.loc[most_used.index, most_used.index]

def style_cells(x):
    helper_df = pd.DataFrame('', index=x.index, columns=x.columns)
    helper_df.loc["time-series", "r"] = "background-color: yellow"
    helper_df.loc["r", "time-series"] = "background-color: yellow"
    for k in range(helper_df.shape[0]):
        helper_df.iloc[k,k] = "color: blue"
    
    return helper_df

relations_most_used.style.apply(style_cells, axis=None)
```

```python
for i in range(relations_most_used.shape[0]):
    relations_most_used.iloc[i,i] = pd.np.NaN
```

```python
relations_most_used
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(relations_most_used, cmap="Greens", annot=False)
```

```python
all_q = pd.read_csv("all_questions.csv", parse_dates=["CreationDate"])
all_q["Tags"] = all_q["Tags"].str.replace("^<|>$", "").str.split("><")
```

```python
def class_deep_learning(tags):
    for tag in tags:
        if tag in ["lstm", "cnn", "scikit-learn", "tensorflow",
                   "keras", "neural-network", "deep-learning"]:
            return 1
    return 0
```

```python
all_q["DeepLearning"] = all_q["Tags"].apply(class_deep_learning)
```

```python
all_q.sample(20)
```

```python
all_q = all_q[all_q["CreationDate"].dt.year < 2020]
```

```python
def fetch_quarter(datetime):
    year = str(datetime.year)[-2:]
    quarter = str(((datetime.month-1) // 3) + 1)
    return "{y}Q{q}".format(y=year, q=quarter)

all_q["Quarter"] = all_q["CreationDate"].apply(fetch_quarter)
```

```python
all_q.head()
```

```python
quarterly = all_q.groupby('Quarter').agg({"DeepLearning": ['sum','size']})
quarterly.columns = ['DeepLearningQuestions', 'TotalQuestions']
quarterly["DeepLearningRate"] = quarterly["DeepLearningQuestions"] / quarterly["TotalQuestions"]
# The following is done to help with visualizations later.
quarterly.reset_index(inplace=True)
quarterly.head(20)
```

```python
ax1 = quarterly.plot(x="Quarter", y="DeepLearningRate",
                    kind="line", linestyle="-", marker="o", color="orange",
                    figsize=(24,12)
                    )

ax2 = quarterly.plot(x="Quarter", y="TotalQuestions",
                     kind="bar", ax=ax1, secondary_y=True, alpha=0.7, rot=45)

for idx, t in quarterly["TotalQuestions"].items():
    ax2.text(idx, t, str(t), ha="center", va="bottom")
    
xlims = ax1.get_xlim()

ax1.get_legend().remove()

# handles1, labels1 = ax1.get_legend_handles_labels()
# handles2, labels2 = ax2.get_legend_handles_labels()
# ax1.legend(handles=handles1 + handles2,
#            labels=labels1 + labels2,
#            loc="upper left", prop={"size": 12})


# for ax in (ax1, ax2):
#     for where in ("top", "right"):
#         ax.spines[where].set_visible(False)
#         ax.tick_params(right=False, labelright=False)
```
